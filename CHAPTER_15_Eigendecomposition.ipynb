```python
import numpy as np
from scipy.linalg import eig,hankel
import matplotlib.pyplot as plt
```

```python
# matrix
A = np.array([[2,3],[3,2]])

# eigendecomposition
L,V = np.linalg.eig(A)

# transform eigenvalues into diagonal matrix
L = np.diag(L)
```

```python
# create two random matrices
n = 3
A = np.random.randn(n,n)
B = np.random.randn(n,n)

# note that this is scipy's eig, not numpy's
evals,evecs = eig(A,B)
```

```python
# initialize results vector
avediffs = np.zeros(100)

# loop over matrix sizes
for n in range(1,101):

    # create matrices
    A = np.random.randn(n,n)
    B = np.random.randn(n,n)

    # GED
    l1 = eig(A,B)[0]
    l2 = eig(np.linalg.inv(B)@A)[0]

    # important to sort eigvals
    l1.sort()
    l2.sort()

    # their differences
    avediffs[n-1] = np.mean(np.abs(l1-l2))


# visualize
plt.plot(avediffs)
plt.show()
```

```python
# create the hankel matrix
t = np.arange(1,51)
lstrow = np.append(t[-1],np.arange(1,t[-1]))
H = hankel(t,r=lstrow)

# eigendecomposition
d,V = np.linalg.eig(H)
V = V[:,np.argsort(d)[::-1]]


# the matrix
plt.subplot(221)
plt.imshow(H)

# the eigenvectors
plt.subplot(222)
plt.imshow(V)

# some evecs
plt.subplot(212)
plt.plot(V[:,:4])

plt.show()
```
